```python
A = 1 # int=inteiro
B = 1.5 #float=decimais
C = "50L" #str=texto
D = 3+2J #complex=complexos
E = True #bool = booleano true ou false
F = "SENAI 2021" #str = texto

type(A)
type(B)
type(C)
type(D)
type(E)
type(F)
```

```python
print ((3*2+1)%2)
print (A != B) #Diferente
print (A == B) #igual
print (A > B) # maior que
print (A < B) # menor que
print (A <= B) # menor ou igual a
print (A >= B) # maior ou igual
```

```python
print(int(3.14)) #converte float para int
print(float(5)) #converte int para float
print(5.0/2+3)
print(5.0/(2+3))
```

```python
c= 3 + 4j
print(c)
print(c.real)
print(c.imag)
print(c.conjugate())
```

```python
a = 'São Paulo'
anos = 38
print("Eu moro no Estado de", a, "há 38 anos.")
print("Eu moro no Estado de "+a+" há 38 anos.")
print("Eu moro no Estado de %s" %a)
print("Eu moro no Estado de", a, "há", anos,"anos.")
```

```python
print("Agora são %2d:%02d h" % (16,30))
print("Porcentagem: %.1f%%" % (5.333)) #.1f uma casa decimal
print("Porcentagem: %.3f%%" % (5.333)) #.3f tres casas decimais

print("Porcentagem: %.1f%% Exponecial: %.2e" % (5.333,0.00314))
print("Decimal: %d - octal: %o - Hexadecimal: %x" % (10,10,10)) 
```

```python
musicos = [("Page","guitarrista","Led Zeppelin"),("Dipsy","Vocalista","Teletubies")]
msg = "{0} é {1} da banda {2}"
for nome, funcao, banda in musicos:
  print(msg.format(nome,funcao,banda))
```

```python
login = [("Daniel", 12345),("João", 54356)] 
msg = "Olá, {0}. sua senha é {1}."

for nome, senha in login:
  print (msg.format(nome,senha))
```
